### Task 2


Secondly, visualise the companies by segregating them according to Market Cap in
three categories: Large Cap(>20,000 crore), Mid Cap (5,000 to 20,000 crore) & 
Small Cap(<5,000).

```python
df1.head()
```

```python
df1['Market Cap']
```

```python
conditions = [
    (df1['Market Cap'] < 5000),
    (df1['Market Cap'] >= 5000) & (df1['Market Cap'] <= 20000),
    (df1['Market Cap'] > 20000)
]

values = ['Small Cap', 'Mid Cap', 'Large Cap']

df1['Categorised Cap'] = np.select(conditions, values)

df1
```

```python
df1['Categorised Cap'].value_counts()
```

```python
plt.figure(figsize = (7,6))
sns.countplot(df1['Categorised Cap'])
```
